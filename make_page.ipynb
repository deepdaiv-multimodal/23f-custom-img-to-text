```python
from PIL import Image, ImageFont, ImageDraw
import numpy as np
import matplotlib.pyplot as plt
```

```python
background = Image.open("input_image/paper.jpeg")
foreground = Image.open("input_image/strawberry_cake.jpeg")
```

```python
box = (200, 200, 1224, 1224)

background.paste(foreground, box)

img_np = np.array(background) ## 행렬로 변환된 이미지
plt.imshow(img_np) ## 행렬 이미지를 다시 이미지로 변경해 디스플레이
plt.show() ## 이미지 인터프린터에 출력
```

```python
#제목을 이미지에
draw = ImageDraw.Draw(background)
draw.font = ImageFont.truetype("/Users/shin-youna/Downloads/maruburi/OTF/MaruBuri-Regular.otf")


def loadfont(fontsize=50):
	# ttf파일의 경로를 지정합니다.
	ttf = '/Users/shin-youna/Downloads/maruburi/OTF/MaruBuri-Regular.otf'
	return ImageFont.truetype(font=ttf, size=fontsize)
    #size : in pixels

titlefontObj = loadfont(50)

title_text = "The Beauty of Mountain"
out_img = ImageDraw.Draw(background)
out_img.text(xy=(1500,100), text=title_text, fill=(237, 230, 211), font=titlefontObj)

img_np = np.array(out_img) ## 행렬로 변환된 이미지
plt.imshow(img_np) ## 행렬 이미지를 다시 이미지로 변경해 디스플레이
plt.show() ## 이미지 인터프린터에 출력
```

```python
# 멀티라인 텍스트 박스 이미지만 생성
# create an image
out = Image.new("RGB", (150, 100), (255, 255, 255))

# get a font
fnt = ImageFont.truetype("/Users/shin-youna/Downloads/maruburi/OTF/MaruBuri-Regular.otf", 40)
# get a drawing context
d = ImageDraw.Draw(out)

# draw multiline text
d.multiline_text((10, 10), "Hello\nWorld", font=fnt, fill=(0, 0, 0))

out.show()
```

```python
# 멀티라인 텍스트를 이미지에 삽입
# get an image
with Image.open("input_image/paper.jpeg").convert("RGBA") as base:

    # make a blank image for he text, initialized to transparent text color
    txt = Image.new("RGBA", base.size, (255, 255, 255, 0))

    # get a font
    fnt = ImageFont.truetype("/Users/shin-youna/Downloads/maruburi/OTF/MaruBuri-Regular.otf", 40)
    # get a drawing context
    d = ImageDraw.Draw(txt)

    # draw text, half opacity
    d.text((1500, 100), "Hello", font=fnt, fill=(255, 255, 255, 128))
    # draw text, full opacity
    d.text((1500, 160), "World", font=fnt, fill=(255, 255, 255, 255))

    out = Image.alpha_composite(base, txt)

    out.show()
```
