```python
#     
# create an image
out = Image.new("RGB", (150, 100), (255, 255, 255))

# get a font
fnt = ImageFont.truetype("/Users/shin-youna/Downloads/maruburi/OTF/MaruBuri-Regular.otf", 40)
# get a drawing context
d = ImageDraw.Draw(out)

# draw multiline text
d.multiline_text((10, 10), "Hello\nWorld", font=fnt, fill=(0, 0, 0))

out.show()
```

```python
#    
# get an image
with Image.open("input_image/paper.jpeg").convert("RGBA") as base:

    # make a blank image for he text, initialized to transparent text color
    txt = Image.new("RGBA", base.size, (255, 255, 255, 0))

    # get a font
    fnt = ImageFont.truetype("/Users/shin-youna/Downloads/maruburi/OTF/MaruBuri-Regular.otf", 40)
    # get a drawing context
    d = ImageDraw.Draw(txt)

    # draw text, half opacity
    d.text((1500, 100), "Hello", font=fnt, fill=(255, 255, 255, 128))
    # draw text, full opacity
    d.text((1500, 160), "World", font=fnt, fill=(255, 255, 255, 255))

    out = Image.alpha_composite(base, txt)

    out.show()
```
